```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data with city + country columns
df_sat = pd.read_csv('../data/satellite/sample_NO2_satellite.csv')
df_ground = pd.read_csv('../data/ground/sample_ground_AQI.csv')

#  Select city + country
selected_city = 'Delhi'
selected_country = 'India'

# Filter satellite and ground data
df_sat_filtered = df_sat[(df_sat['city'] == selected_city) & (df_sat['country'] == selected_country)]
df_ground_filtered = df_ground[(df_ground['city'] == selected_city) & (df_ground['country'] == selected_country)]

# Merge by date
df = pd.merge(df_sat_filtered, df_ground_filtered, on='date')

# Plot
plt.figure(figsize=(10, 5))
sns.lineplot(x='date', y='NO2_level', data=df, label='Satellite NO', marker='o')
sns.lineplot(x='date', y='AQI', data=df, label='Ground AQI', marker='o')
plt.title(f'Air Pollution Trend in {selected_city}, {selected_country}')
plt.xlabel('Date')
plt.ylabel('Pollution Level')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
